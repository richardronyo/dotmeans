In this notebook I will be Clustering the NBA Draft Combine participants based on their measurements.

Foremost, I will use 5 clusters to see if the measurements will determine the players position. The measurements I will use are height weight, wingspan, standing vert, running vert, and sprint.

```python
from kmeans import KMeans
import pandas as pd
import numpy as np

np.set_printoptions(threshold=np.inf)
COMBINE = pd.read_csv(("data/NBA_draft_combine.csv"))

#Getting the appropriate columns from our data and getting rid of all the empty rows
names_and_measurements = COMBINE[["PLAYER", "POS", "HGT", "WGT", "WNGSPN", "STNDVERT", "LPVERT", "SPRINT"]].dropna()
D = names_and_measurements.to_numpy()
print(D)

D[:, 3:] = pd.to_numeric(D[:, 3:], errors = "coerce")
D[:, 3:] = D[:, 3:].astype(float)
print(D)

D = D[~np.isnan(D).any(axis=1)]
names_and_pos = names_and_measurements[["PLAYER", "POS"]]
physical_cluster = KMeans(D, 3)
physical_cluster.plot_clustered_data()
```
